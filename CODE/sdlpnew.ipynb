```python
import numpy as np 
import pandas as pd 
import types

import plotly.plotly as py
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go

from wordcloud import WordCloud

import matplotlib.pyplot as plt

import os
print(os.listdir(r"C:\Users\KAUSTUBH\Desktop"))
```

```python
countryData = pd.read_csv("C:\Users\KAUSTUBH\Desktop\countries of the world.csv")
```

```python
countryData.info()
```

```python
countryData.drop('Phones (per 1000)', axis=1, inplace=True)
countryData.drop('Coastline (coast/area ratio)', axis=1, inplace=True)
countryData.drop('Crops (%)', axis=1, inplace=True)
countryData.drop('Other (%)', axis=1, inplace=True)
countryData.drop('Climate', axis=1, inplace=True)

countryData['Pop. Density (per sq. mi.)']=countryData['Pop. Density (per sq. mi.)'].str.replace(',','.')
countryData['Pop. Density (per sq. mi.)']=countryData['Pop. Density (per sq. mi.)'].astype('float')

countryData['Net migration']=countryData['Net migration'].str.replace(',','.')
countryData['Net migration']=countryData['Net migration'].astype('float')

countryData['Infant mortality (per 1000 births)']=countryData['Infant mortality (per 1000 births)'].str.replace(',','.')
countryData['Infant mortality (per 1000 births)']=countryData['Infant mortality (per 1000 births)'].astype('float')

countryData['Literacy (%)']=countryData['Literacy (%)'].str.replace(',','.')
countryData['Literacy (%)']=countryData['Literacy (%)'].astype('float')

countryData['Arable (%)']=countryData['Arable (%)'].str.replace(',','.')
countryData['Arable (%)']=countryData['Arable (%)'].astype('float')

countryData['Birthrate']=countryData['Birthrate'].str.replace(',','.')
countryData['Birthrate']=countryData['Birthrate'].astype('float')

countryData['Deathrate']=countryData['Deathrate'].str.replace(',','.')
countryData['Deathrate']=countryData['Deathrate'].astype('float')

countryData['Agriculture']=countryData['Agriculture'].str.replace(',','.')
countryData['Agriculture']=countryData['Agriculture'].astype('float')

countryData['Industry']=countryData['Industry'].str.replace(',','.')
countryData['Industry']=countryData['Industry'].astype('float')

countryData['Service']=countryData['Service'].str.replace(',','.')
countryData['Service']=countryData['Service'].astype('float')
```

```python
countryData = countryData.fillna(0)
```

```python
africaData = countryData[countryData["Region"].str.contains('AFRICA')]
africaData = africaData.sort_values(["GDP ($ per capita)"],ascending = False)

namericaData = countryData[countryData["Region"].str.contains('NORTHERN AMERICA')]
namericaData = namericaData.sort_values(["GDP ($ per capita)"],ascending = False)

lamericaData = countryData[countryData["Region"].str.contains('LATIN AMER. & CARIB')]
lamericaData = lamericaData.sort_values(["GDP ($ per capita)"],ascending = False)

oceaniaData = countryData[countryData["Region"].str.contains('OCEANIA')]
oceaniaData = oceaniaData.sort_values(["GDP ($ per capita)"],ascending = False)

asiaData = countryData[countryData["Region"].str.contains('NEAR EAST|C.W. OF IND. STATES')]
asiaData = asiaData.sort_values(["GDP ($ per capita)"],ascending = False)

europeData = countryData[countryData["Region"].str.contains('EUROPE|BALTICS')]
europeData = europeData.sort_values(["GDP ($ per capita)"],ascending = False)
```
